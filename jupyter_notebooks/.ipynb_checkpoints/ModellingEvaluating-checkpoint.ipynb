# Image Data Augmentation
The dataset can be artificially expanded by augmenting images using the Keras class ImageDataGenerator. This class enables configuration of random transformations and normalization operations to be done on the image data during training and generation of batches of additional image data and their labels directly from the jpgs in their respective folders.

## Import ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Initialize ImageDataGenerator

The ImageDataGenerator function will be assigned to the `augmented_image_data` variable and intialized to perform the following augmentations:

* Randomly rotate images within a range of 0-40 degrees.
* Randomly translate images vertically by up to 10% of the total image height.
* Randomly translate images horizontally by up to 10% of the total image width.
* Randomly apply shearing transformations to distort the image along an axis, to a degree of 10%.
* Randomly zoom inside images to a degree of 10%.
* Randomly flip images horizontally.
* Randomly flip images vertically.
* Newly created or broken pixels, which can appear after a rotation or a width/height shift, will be filled in with the nearset pixel value.
* Pixel values will be rescaled to target values between 0 and 1 by scaling with a 1/255. factor. This helps to avoid pixel values that are too high for the model to process.

```python
augmented_image_data = ImageDataGenerator(rotation_range=40,
                                          height_shift_range=0.10,
                                          width_shift_range=0.10,
                                          shear_range=0.1,
                                          zoom_range=0.1,
                                          horizontal_flip=True,
                                          vertical_flip=True,
                                          fill_mode='nearest',
                                          rescale=1./255
                                          )
```

## Augment Training Image Dataset

Apply `augmented_image_data` to the training set using the `flow_from_directory` function to fetch images directly from the training folder, transform each image width and height to the target image shape set during data analysis, set images to standard RBG color mode, set the batch number of images to augment, set `class_mode` to binary as we have two classes of images and shuffle the images. The resulting values will be assigned to the variable `train_set`.

Class indices are displayed to verify that images from both classes were augmented.

```python
batch_size = 20
train_set = augmented_image_data.flow_from_directory(train_path,
                                                     target_size=image_shape[:2],
                                                     color_mode='rgb',
                                                     batch_size=batch_size,
                                                     class_mode='categorical',
                                                     shuffle=True
                                                     )

train_set.class_indices
```

## Augment Validation and Test Image Datasets
Fetch images from the validation and test image datasets using `flow_from_directory` function with the same arguments used to fetch images from the train dataset, however the images will be augmented only by standardizing image pixels between 0 to 255 pixels using `ImageDataGenerator`.

Class indices are displayed to verify that images from both classes were augmented.

### Augment Validation Image Dataset

```python
validation_set = ImageDataGenerator(rescale=1./255).flow_from_directory(val_path,
                                                                        target_size=image_shape[:2],
                                                                        color_mode='rgb',
                                                                        batch_size=batch_size,
                                                                        class_mode='categorical',
                                                                        shuffle=False
                                                                        )

validation_set.class_indices
```

### Augment Test Image Dataset

```python
test_set = ImageDataGenerator(rescale=1./255).flow_from_directory(test_path,
                                                                  target_size=image_shape[:2],
                                                                  color_mode='rgb',
                                                                  batch_size=batch_size,
                                                                  class_mode='categorical',
                                                                  shuffle=False
                                                                  )

test_set.class_indices
```

# Model Creation

```python
!pip install keras-tuner -q
```

Import Model Packages

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
import keras_tuner
```

## Define the Search Space

Create a hypermodel with dynamic hyperparameters. A model is defined with a convolution layer as the input for which the number of filters will be tuned. The number of subsequent hidden layers will be tuned, these will be convolution layers for which the number of filters will also be tuned. These layers are followed by a flatten layer and then a dense layer for which the number of units will be tuned. Next a dropout layers for which the drop out rate will be tuned. The model finishes with a dense layer and is compiled with a choosen optimizer. This process will enable testing and comparrison of the use of different model architectures and hyperparameters.

```python
def build_model(hp):
    model = Sequential()

    model.add(Conv2D(hp.Int('input_filters', 32, 64, step=32),
                     kernel_size=(3, 3),
                     input_shape=image_shape,
                     activation='relu', ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    for i in range(hp.Int('n_layers', 1, 3)):
        model.add(Conv2D(hp.Int(f'conv_{i}_filters', 32, 64, step=32),
                         kernel_size=(3, 3),
                         activation='relu', ))
        model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(hp.Int("units", min_value=64, max_value=128, step=64),
                    activation='relu'))
    
    model.add(Dropout(hp.Float("dropout_rate", min_value=0.2, max_value=0.5, step=0.3)))
    model.add(Dense(2, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    return model
```

```python
build_model(keras_tuner.HyperParameters())
```

## Start the Search

With the search space defined, a tuner class is initialized to run the search.

```python
tuner = keras_tuner.RandomSearch(
    hypermodel=build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=1,
    overwrite=True,
    directory=f'{file_path}',
    project_name='tuner_results',
)
```

### Print a Summary of the Search Space

```python
tuner.search_space_summary()
```

### Define Callbacks

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, TensorBoard

early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1, min_lr=0.001)
tb = TensorBoard(f'{file_path}/tb_logs')
```

### Perform Search for Best Hyperparameter Configuration

To find the best set of hyperparameter values, the model-building function is called with different hyperparameter values for each trial. The model is fit, evaluated and the metrics are recorded.

All the arguments passed to search are passed to `model.fit()` in each execution.

```python
tuner.search(train_set,
             epochs=10,
             validation_data=validation_set,
             callbacks=[early_stop, reduce_lr, tb],
)
```

## View the Hyperparameter Tuning Process in Tensorboard

The hyperparamter tuning process was viewed using Tensorboard (detailed in README.md).

# Retrain the Model using the Optimized Hyperparameter Values

```python
def build_opt_model():
    model = Sequential()

    model.add(Conv2D(64,
                     kernel_size=(3, 3),
                     input_shape=image_shape,
                     activation='relu', ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(32,
                     kernel_size=(3, 3),
                     input_shape=image_shape,
                     activation='relu', ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(32,
                     kernel_size=(3, 3),
                     input_shape=image_shape,
                     activation='relu', ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(32,
                     kernel_size=(3, 3),
                     input_shape=image_shape,
                     activation='relu', ))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(128,
                    activation='relu'))
    
    model.add(Dropout(0.2))
    model.add(Dense(2, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    return model
```

## Model Summary

```python
build_opt_model().summary()
```

## Fit Model for Model Training

```python
model = build_opt_model()
model.fit(train_set,
          epochs=10,
          validation_data=validation_set,
          callbacks=[early_stop, reduce_lr]
         )
```

## Save Model

```python
model.save('outputs/v2/mildew_detector_model.h5')
```
